```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

citra = cv2.imread('C:/Users/USER/Downloads/PROJECT FIX/DATASET_grayscale(3)/A/A1.jpg', cv2.IMREAD_GRAYSCALE)

h1=0
h2=1
h3=0
h4=1
h5=3
h6=1
h7=0
h8=1
h9=0

m, n = citra.shape
f = np.double(citra)
G8 = np.zeros((m,n))

for x in range(1, m-1):
    for y in range(1, n-1):
        G8[x,y] = 1/7*(h1*f[x-1,y-1] + h2*f[x-1,y] + h3*f[x-1,y-1] + \
                        h4*f[x,y-1] + h5*f[x,y] + h6*f[x,y+1] + \
                        h7*f[x+1,y-1] + h8*f[x+1,y] + h9*f[x+1,y+1])
        
G8 = np.uint8(G8)

plt.subplot(1,2,1), plt.imshow(citra, cmap='gray')
plt.title('Citra Asli')
plt.subplot(1,2,2), plt.imshow(G8, cmap='gray')
plt.title('Gaussian Blur')
plt.show()
cv2.imwrite('C:/Users/USER/Downloads/PROJECT FIX/DATASET_gaussian_blur(3)/A/A1.jpg',G8)
```

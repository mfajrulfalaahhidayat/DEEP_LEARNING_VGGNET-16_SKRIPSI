```python
from matplotlib.pyplot import imread, imshow, subplot, title
import cv2
import numpy as np

citra = cv2.imread('C:/Users/USER/Downloads/PROJECT FIX/DATASET_gaussian_blur(3)/A/A1.jpg', cv2.IMREAD_GRAYSCALE)

m, n = citra.shape
ambang = np.zeros((m, n))
G = np.zeros((m, n), dtype=np.uint8)
f = citra.astype('int')
C = 3

for x in range(4, m-4):
    for y in range(4, n-4):
#         jum = f[x-1,y-1] + f[x-1,y] + f[x-1,y+1] + \
#             f[x,y-1] + f[x,y] + f[x,y+1] + \
#             f[x+1,y-1] + f[x+1,y] + f[x+1,y+1]
        
        jum = f[x-4,y-4] + f[x-4,y-3] + f[x-4,y-2] + f[x-4,y-1] + f[x-4,y] + f[x-4,y+1] + f[x-4,y+2] + f[x-4,y+3] + f[x-4,y+4] + \
              f[x-3,y-4] + f[x-3,y-3] + f[x-3,y-2] + f[x-3,y-1] + f[x-3,y] + f[x-3,y+1] + f[x-3,y+2] + f[x-3,y+3] + f[x-3,y+4] + \
              f[x-2,y-4] + f[x-2,y-3] + f[x-2,y-2] + f[x-2,y-1] + f[x-2,y] + f[x-2,y+1] + f[x-2,y+2] + f[x-2,y+3] + f[x-2,y+4] + \
              f[x-1,y-4] + f[x-1,y-3] + f[x-1,y-2] + f[x-1,y-1] + f[x-1,y] + f[x-1,y+1] + f[x-1,y+2] + f[x-1,y+3] + f[x-1,y+4] + \
              f[x,y-4]   + f[x,y-3]   + f[x,y-2]   + f[x,y-1]   + f[x,y]   + f[x,y+1]   + f[x,y+2]   + f[x,y+3]   + f[x,y+4]   + \
              f[x+1,y-4] + f[x+1,y-3] + f[x+1,y-2] + f[x+1,y-1] + f[x+1,y] + f[x+1,y+1] + f[x+1,y+2] + f[x+1,y+3] + f[x+1,y+4] + \
              f[x+2,y-4] + f[x+2,y-3] + f[x+2,y-2] + f[x+2,y-1] + f[x+2,y] + f[x+2,y+1] + f[x+2,y+2] + f[x+2,y+3] + f[x+2,y+4] + \
              f[x+3,y-4] + f[x+3,y-3] + f[x+3,y-2] + f[x+3,y-1] + f[x+3,y] + f[x+3,y+1] + f[x+3,y+2] + f[x+3,y+3] + f[x+3,y+4] + \
              f[x+4,y-4] + f[x+4,y-3] + f[x+4,y-2] + f[x+4,y-1] + f[x+4,y] + f[x+4,y+1] + f[x+4,y+2] + f[x+4,y+3] + f[x+4,y+4] 
        ambang[x,y] = np.uint8(1/81 * jum) - C
        
        if ambang[x,y] > f[x,y]:
            G[x,y] = 0
        else:
            G[x,y] = 255


subplot(1,2,1)
imshow(citra, cmap='gray')
title('Citra Asli')

subplot(1,2,2)
imshow(G, cmap='gray')
title('Adaptive Mean Thresholding')

# cv2.imwrite('C:/Users/USER/Downloads/PROJECT FIX/DATASET_adaptive_mean_thresholding(3)/A/A1.jpg',G)
```

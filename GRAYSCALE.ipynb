```python
# Import library yang diperlukan
# import numpy as np

def rgb2gray(img):
    # Memisahkan channel R, G, dan B dari gambar
    red = img[:,:,0]
    green = img[:,:,1]
    blue = img[:,:,2]
    
    # Menghitung nilai gray dengan mengalikan masing-masing channel dengan bobot yang sesuai
    gray = (0.2126 * red) + (0.7152 * green) + (0.0722 * blue)
    
    # Mengembalikan nilai gray
    return gray
```
